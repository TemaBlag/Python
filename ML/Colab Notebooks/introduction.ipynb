```python
import numpy as np
def cumulative_mean(X):
    result = np.cumsum(X)
    new_arr = np.arange(1,len(X) + 1)

    result = result / new_arr
    return result
# test
a = np.array([1, 2, 4, -9])
print(cumulative_mean(a))

assert np.allclose(cumulative_mean(a), np.array([1, 1.5, 7./3, -0.5]))
```

```python
import numpy as np
def cumulative_mean_2d(A):
    result =  np.cumsum(A, axis=1)
    new_arr = np.cumsum(np.ones(A.shape), axis=1)
    result = result / new_arr
    return result
print(cumulative_mean_2d(np.array([[1, 2, 3],[4, 5, 6]])))
```

```python
def diag_2k(a):
    sum = 0
    for i in np.diag(a):
        sum += 0 if i % 2 else i
    return sum
```

```python
diag_2k([[2, 2, 3], [4, 5, 6], [7, 8, 9]])
```

```python
import numpy as np

def calculate_loss_function(X, w, y):
    return np.square(np.linalg.norm(X @ w - y.T))
X = np.array([[1, 2, 3], [4, 5, 6]])
w = np.array([1, 2, 3])
y = np.array([1, 2])
print(calculate_loss_function(X, w, y))
```

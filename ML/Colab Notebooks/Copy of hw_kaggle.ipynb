```python
d = {}
def chek_not_numeric(x):
  try:
    x = float(x)
    return x
  except ValueError:
    if x not in d:
      d[x] = 1
    else:
      d[x] += 1
    return np.nan


data["TotalSpent"] = data["TotalSpent"].apply(chek_not_numeric)
print(d)
del d
```

(более сложный вариант) Если вы будете использовать кроссвалидацию, то преобразования данных и LogisticRegression нужно соединить в один Pipeline с помощью make_pipeline, как это делалось во втором семинаре. Потом pipeline надо передать в GridSearchCV. Для one-hot-encoding'a можно испльзовать комбинацию LabelEncoder + OneHotEncoder (сначала превращаем строчки в числа, а потом числа првращаем в one-hot вектора.)

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.compose import ColumnTransformer
from sklearn import metrics
from sklearn.impute import SimpleImputer
```

```python
num_preproc = make_pipeline(SimpleImputer(strategy="median"),StandardScaler())
preproc = ColumnTransformer([
    ("num",num_preproc,num_cols),
    ("cat",OneHotEncoder(handle_unknown='ignore',sparse_output=False),cat_cols)
])
pipeline = Pipeline([("preproc",preproc),
                     ("model" ,LogisticRegression())])
```

```python
params ={
    "model__C":[100, 10, 1, 0.1, 0.01, 0.001]
}
```

```python
grid = GridSearchCV(pipeline,params,scoring='roc_auc',n_jobs=-1)
grid.fit(data.drop(columns=target_col),data[target_col])
grid.best_params_, grid.best_score_
```

```python
best_C = grid.best_params_["model__C"]
pipeline.set_params(model__C = best_C )
```

```python
pipeline.fit(data.drop(columns=target_col),data[target_col])
```

```python
from catboost import CatBoostClassifier, Pool, cv
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    data.drop(columns=target_col), data[target_col], test_size=0.33)
```

```python
train_dataset = Pool(data=X_train,
                     label=y_train,
                     cat_features=cat_cols)
test_dataset = Pool(data=X_test,
                     label=y_test,
                     cat_features=cat_cols)
```

```python
its = [100,500,1000]
boosting_metrics = []
lrs = [0.001,0.1,0.03,0.01]
top = (-1,0,0) # scor, lerning rate, depth
for lr in lrs:
  for  it in its:
    params = {"iterations": it,
              "learning_rate" : lr,
              "loss_function": "Logloss",
              "custom_loss": "AUC",
              "verbose": False}

    scores = cv(pool=train_dataset,params=params,fold_count=5,verbose=False) #ВАШ КОД. Используйте функцию cv
    scor = scores["test-AUC-mean"][99]
    print(f"scor={scor}, lr={lr}, it={it}")
    if scor > top[0]:
      top = (scor,lr,it)
    boosting_metrics.append(scores)
top
```

```python
top
```

```python
rf = CatBoostClassifier(learning_rate=0.1,iterations=500)
rf.fit(train_dataset)
```

```python
roc_auc_score(y_test,rf.predict_proba(X_test)[:,1])
```

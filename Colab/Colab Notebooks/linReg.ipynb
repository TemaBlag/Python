```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('penguins.csv')
df.head()
```

```python
df = df.rename(columns={
    'вид' : 'kind',
    'остров' : 'island',
    'длина клюва' : 'beak_length',
    'ширина клюва' : 'beak_width',
    'длина плавника' : 'fin_length',
    'масса тела' : 'weight',
    'пол' : 'sex'
})
```

```python
nullData = df.isna()
```

```python
df.head()
```

```python
df.describe()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[['weight']]
Y = df['fin_length']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15)
```

```python
plt.scatter(X_train, Y_train)
plt.xlabel('Weight penguin')
plt.ylabel('Fin length')
plt.show()
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
Y_predict_train = lr.predict(X_train)
Y_predict_test = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
np.array(Y_train)
```

```python
Y_predict_train
```

```python
a = lr.coef_
```

```python
b = lr.intercept_
```

```python
plt.scatter(X_train, Y_train)

def lin(x):
  return a * x + b

x = [2500, 6500]
plt.plot(x, lin(x))
plt.xlabel('Weight penguin')
plt.ylabel('Fin length')
plt.show()
```

```python
plt.scatter(X_test, Y_test)

def f_line(x):
  return  lr.coef_*x + lr.intercept_

x_line = [2500, 6000]
y_line = f_line(x_line)
plt.plot(x_line, y_line)

plt.xlabel('масса тела (гр)')
plt.ylabel('длина плавника (мм)')

plt.show()
```

```python
np.array(Y_test)[:10], Y_predict_test[:10]
```

```python
mean_squared_error(Y_test, Y_predict_test), mean_absolute_error(Y_test, Y_predict_test)
```

```python
mean_squared_error(Y_train, Y_predict_train), mean_absolute_error(Y_train, Y_predict_train)
```

```python
np.mean(np.array(abs(Y_test-Y_predict_test)))
```

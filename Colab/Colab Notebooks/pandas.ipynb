```python
import pandas as pd

df_rest = pd.read_csv('https://ai-academy.ru/upload/csv/Restaurant.csv')
df_rest['vegan'] = df_rest['vegan'].replace({'No': 0, 'Yes': 1})
df_rest.head()
```

```python
data = df_rest.pivot_table(values=['total_bill','vegan'],
                           index='day')
data.head()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
fig = plt.figure()
axes=fig.add_axes([0, 0, 1, 1])
axes.bar(x = data.index, height = data['total_bill'],
       width=0.4, align='edge', label = 'Средняя сумма чека')
axes.bar(x=data.index, height = data['vegan'],
       width=-0.4, align='edge', label = 'Веган')
axes.set_title("Средняя сумма счёта Vs vegan")
axes.set_xlabel('Дни недели')
axes.set_ylabel('Cyммa')
```

```python
axes = df_rest.plot.scatter("total_bill", "vegan")
```

```python
axes = df_rest.plot.scatter("total_bill", "tip", c='DarkBlue')

axes.set_title('Общая сумма счёта Vs Cумма чаевых')
axes.set_xlabel('Сумма чека')
axes.set_ylabel('Размер чаевых')
```

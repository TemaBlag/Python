```python
class Neuron:

    def __init__(self, w, f = lambda x: x):
        self.neurons = w
        self.func = f 
        self.mas_x = None

    def forward(self, x):
        sum1 = 0
        self.mas_x = x
        for i in range(len(self.mas_x)):
            sum1 += (self.mas_x[i] * self.neurons[i])
        return self.func(sum1)

    def backlog(self):
        return self.mas_x

    
neuron = Neuron([1,2,3])
print(neuron.forward([4,4,4]))


```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import hypergeom

# Параметры гипергеометрического распределения
N = 56  # размер генеральной совокупности
K = 36 # количество успехов в генеральной совокупности
n = 2  # размер выборки
# Создание объекта гипергеометрического распределения
dist = hypergeom(N, K, n)

x = np.arange(0, n+1)
pmf_que = dist.pmf(x)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, pmf_que, 'bo')
ax.vlines(x, 0, pmf_que, lw=2)
ax.set_xlabel('# of learned questions in our group of all')
ax.set_ylabel('hypergeom PMF')

plt.show()
```
